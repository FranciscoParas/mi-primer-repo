```python
import matplotlib.pyplot as plt
```

```python
# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]

# Crear el gráfico de dispersión
plt.scatter(x, y, color='blue', marker='o')

# Añadir títulos y etiquetas
plt.title('Gráfico de dispersión simple')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Mostrar el gráfico
plt.show()
```
